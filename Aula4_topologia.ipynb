# Curvatura de Berry e número de Chern — intuição e o método de Fukui

## 1) Fase geométrica (Berry)
Num cristal periódico,
$$
|\psi_{n\mathbf{k}}\rangle = e^{i\mathbf{k}\cdot\mathbf{r}}\,|u_{n\mathbf{k}}\rangle,
$$
onde $|u_{n\mathbf{k}}\rangle$ é periódico na rede. Ele não é único: podemos trocar a fase local (gauge)
$$
|u_{n\mathbf{k}}\rangle \rightarrow e^{i\phi(\mathbf{k})}\,|u_{n\mathbf{k}}\rangle
$$
sem mudar observáveis.

A conexão de Berry é
$$
\mathbf{A}_n(\mathbf{k}) = i\langle u_{n\mathbf{k}}|\nabla_{\mathbf{k}}u_{n\mathbf{k}}\rangle,
$$
e a curvatura de Berry em 2D é
$$
\Omega_n(\mathbf{k}) = \left(\nabla_{\mathbf{k}}\times \mathbf{A}_n(\mathbf{k})\right)_z.
$$

**Intuição:** $\mathbf{A}$ faz o papel de “potencial vetor” e $\Omega$ de “campo magnético” no espaço-$k$.
Regiões de avoided crossings (gap pequeno) tendem a concentrar $\Omega$.

---

## 2) Por que o Chern é inteiro?
O número de Chern da banda $n$ em 2D é
$$
C_n=\frac{1}{2\pi}\int_{\text{BZ}} d^2k\,\Omega_n(\mathbf{k}).
$$

Em um cristal 2D, $k_x$ e $k_y$ são periódicos, então a zona de Brillouin (BZ) é topologicamente um **toro** $T^2$.
Quando $C\neq 0$, não dá para escolher uma fase $\phi(\mathbf{k})$ global e suave em todo o toro: existe uma obstrução
global, e $C$ mede isso.

**Comentário (meios contínuos):** em alguns problemas de ótica/topologia em meios contínuos, o domínio natural pode ser
o plano $\mathbb{R}^2$ em vez do toro. Para obter um invariante quantizado, entra uma compactificação/regularização
(o comportamento em $|\mathbf{k}|\to\infty$ passa a importar). Em cristais, o espaço de momentos já é compacto (a BZ).

---

## 3) Fukui: Chern em uma malha discreta (gauge-invariante)
Calcular $\mathbf{A}$ por derivadas numéricas pode ser instável porque a fase de $|u\rangle$ é arbitrária.
O método de Fukui evita derivadas e usa apenas produtos internos entre autovetores vizinhos.

Em uma malha discreta de pontos $\mathbf{k}$, definimos os **links** (para uma banda isolada):
$$
U_x(\mathbf{k})=
\frac{\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_x)\rangle}
{\left|\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_x)\rangle\right|},
\qquad
U_y(\mathbf{k})=
\frac{\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_y)\rangle}
{\left|\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_y)\rangle\right|}.
$$
Eles são fases puras ($U\in U(1)$) e são robustos contra mudanças de gauge.

O “fluxo” discreto em um plaquete é:
$$
F_{xy}(\mathbf{k})=
\operatorname{Im}\ln\Big(
U_x(\mathbf{k})\,
U_y(\mathbf{k}+\Delta k_x)\,
U_x(\mathbf{k}+\Delta k_y)^{-1}\,
U_y(\mathbf{k})^{-1}
\Big),
$$
onde escolhemos o ramo do log de modo que $F_{xy}\in(-\pi,\pi]$.

O Chern number é a soma do fluxo em todos os plaquetes:
$$
C=\frac{1}{2\pi}\sum_{\mathbf{k}} F_{xy}(\mathbf{k}).
$$

---

## 4) Chern de qualquer operador Hermitiano (e o que isso significa)
Matematicamente, podemos definir um Chern number para o conjunto de autovetores de qualquer operador Hermitiano
$O(\mathbf{k})$ suave em $\mathbf{k}$, desde que exista um subespaço bem separado (um “gap” espectral).
Isso é um fato geométrico sobre o *bundle* de autovetores.

Mas o significado físico depende do operador. Para respostas de transporte (como Hall quantizado) e bulk–boundary,
o Chern relevante é o do Hamiltoniano (ou do projetor sobre os estados ocupados do Hamiltoniano), porque é ele que
governa a dinâmica e a ocupação.

## Exemplo:

### 1) O modelo (Qi–Wu–Zhang / Chern insulator)
Vamos usar um Hamiltoniano mínimo de duas bandas (um “pseudospin” $2\times 2$), escrito como
$$
H(\mathbf{k}) = d_x(\mathbf{k})\,\sigma_x + d_y(\mathbf{k})\,\sigma_y + d_z(\mathbf{k})\,\sigma_z,
$$
com
$$
d_x=\sin k_x,\qquad d_y=\sin k_y,\qquad d_z = m+\cos k_x+\cos k_y.
$$

- $\sigma_{x,y,z}$ são matrizes de Pauli no espaço das duas bandas (não é spin real aqui).
- $m$ é um parâmetro de “massa” que controla a fase topológica.
- O espectro é
$$
E_\pm(\mathbf{k}) = \pm |\mathbf{d}(\mathbf{k})|.
$$
A banda inferior $E_-(\mathbf{k})$ é a “ocupada” no caso half-filling.

**Intuição física:** ao variar $\mathbf{k}$ na zona de Brillouin, o autovetor da banda ocupada define uma textura no espaço interno.
A curvatura de Berry mede a rotação local desse autovetor em $\mathbf{k}$-space, e o número de Chern mede o “fluxo total” de curvatura
pela BZ.

---

### 2) Curvatura de Berry e número de Chern
Para a banda ocupada, definimos a curvatura de Berry $\Omega(\mathbf{k})$ e o número de Chern
$$
C=\frac{1}{2\pi}\int_{\mathrm{BZ}} d^2k\,\Omega(\mathbf{k}).
$$
Quando o gap não fecha, $C$ é quantizado e não muda sob pequenas deformações do Hamiltoniano.

---

### 3) Como calculamos numericamente: método de Fukui (versão discreta e robusta)
Em uma malha discreta de pontos $(k_x,k_y)$, calculamos o autovetor normalizado $|u(\mathbf{k})\rangle$ da banda ocupada.
Em vez de derivadas numéricas (sensíveis à escolha de fase), usamos links gauge-invariantes definidos por sobreposições entre
autovetores vizinhos:

$$
U_x(\mathbf{k})=\frac{\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_x)\rangle}
{|\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_x)\rangle|},\qquad
U_y(\mathbf{k})=\frac{\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_y)\rangle}
{|\langle u(\mathbf{k})|u(\mathbf{k}+\Delta k_y)\rangle|}.
$$

O fluxo discreto (curvatura discreta) em cada plaquete é
$$
F_{xy}(\mathbf{k})=\operatorname{Im}\ln\Big(
U_x(\mathbf{k})\,
U_y(\mathbf{k}+\Delta k_x)\,
U_x(\mathbf{k}+\Delta k_y)^{-1}\,
U_y(\mathbf{k})^{-1}
\Big),
$$
e o Chern number é
$$
C=\frac{1}{2\pi}\sum_{\mathbf{k}} F_{xy}(\mathbf{k}).
$$

**Por que é robusto:** tudo é construído a partir de fases de produtos internos, que são invariantes sob mudanças de gauge
$|u(\mathbf{k})\rangle\to e^{i\phi(\mathbf{k})}|u(\mathbf{k})\rangle$.

---

### 4) Testes que vamos fazer (para checar que está tudo funcionando)

1. **Verificar que o gap não fecha** (o método só faz sentido se a banda ocupada estiver bem separada).
   No código, imprimimos o menor gap $E_+-E_-$ na malha.

2. **Verificar a quantização do Chern**: o resultado deve ser muito próximo de um inteiro e ficar mais estável
   ao aumentarmos a malha $(N_{kx},N_{ky})$.

3. **Varredura em $m$ (fases esperadas)**  
   Experimente valores típicos (com a banda inferior ocupada):
   - $m=-3$  $\Rightarrow$ fase trivial, $C\approx 0$
   - $m=-1$  $\Rightarrow$ fase topológica, $C\approx +1$
   - $m=+1$  $\Rightarrow$ fase topológica, $C\approx -1$
   - $m=+3$  $\Rightarrow$ fase trivial, $C\approx 0$

4. **Visualizar $F_{xy}(k)$**: o mapa de $F_{xy}$ mostra onde a curvatura se concentra na BZ (tipicamente perto de regiões onde
   o gap é menor), ajudando a criar intuição geométrica.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA

# -------------------------
# Modelo 2-band (QWZ)
# -------------------------
sx = np.array([[0, 1], [1, 0]], dtype=complex)
sy = np.array([[0, -1j], [1j, 0]], dtype=complex)
sz = np.array([[1, 0], [0, -1]], dtype=complex)
I2 = np.eye(2, dtype=complex)

def Hk(kx, ky, m=-1.0):
    dx = np.sin(kx)
    dy = np.sin(ky)
    dz = m + np.cos(kx) + np.cos(ky)
    return dx*sx + dy*sy + dz*sz

# -------------------------
# Fukui (banda única)
# -------------------------
def _normalize_phase(z, eps=1e-14):
    a = np.abs(z)
    return z/a if a > eps else 1.0 + 0.0j

def fukui_chern_single_band(u):
    Nkx, Nky, dim = u.shape
    Ux = np.zeros((Nkx, Nky), dtype=complex)
    Uy = np.zeros((Nkx, Nky), dtype=complex)

    for ix in range(Nkx):
        ixp = (ix + 1) % Nkx
        for iy in range(Nky):
            iyp = (iy + 1) % Nky
            ovx = np.vdot(u[ix, iy], u[ixp, iy])
            ovy = np.vdot(u[ix, iy], u[ix, iyp])
            Ux[ix, iy] = _normalize_phase(ovx)
            Uy[ix, iy] = _normalize_phase(ovy)

    Fxy = np.zeros((Nkx, Nky), dtype=float)
    for ix in range(Nkx):
        ixp = (ix + 1) % Nkx
        for iy in range(Nky):
            iyp = (iy + 1) % Nky
            W = Ux[ix, iy] * Uy[ixp, iy] / (Ux[ix, iyp] * Uy[ix, iy])
            Fxy[ix, iy] = np.angle(W)  # Im log no ramo principal

    C = np.sum(Fxy) / (2*np.pi)
    return C, Fxy

# -------------------------
# Cálculo numérico: autovetores, curvatura discreta e Chern
# -------------------------
Nkx = Nky = 61
kx_list = np.linspace(-np.pi, np.pi, Nkx, endpoint=False)
ky_list = np.linspace(-np.pi, np.pi, Nky, endpoint=False)

m = -1.0  # experimente: m=-3 (C=0), m=-1 (C=+1), m=+1 (C=-1), m=+3 (C=0)
band = 0  # banda ocupada (a mais baixa)

u = np.zeros((Nkx, Nky, 2), dtype=complex)
E = np.zeros((Nkx, Nky, 2), dtype=float)

gap_min = 1e9
for ix, kx in enumerate(kx_list):
    for iy, ky in enumerate(ky_list):
        H = Hk(kx, ky, m=m)
        evals, evecs = LA.eigh(H)
        E[ix, iy] = evals.real
        u[ix, iy] = evecs[:, band]
        gap_min = min(gap_min, (evals[1] - evals[0]).real)

C, Fxy = fukui_chern_single_band(u)
print(f"m = {m: .3f} | Chern (banda {band}) = {C:.6f} | gap mínimo ~ {gap_min:.6f}")

# -------------------------
# Visualização: Berry curvature discreta (fluxo por plaquete)
# -------------------------
plt.figure(figsize=(6.5, 4.8))
plt.imshow(
    Fxy.T, origin="lower", aspect="auto",
    extent=[kx_list[0], kx_list[-1], ky_list[0], ky_list[-1]]
)
plt.xlabel(r"$k_x$")
plt.ylabel(r"$k_y$")
plt.title(r"Fukui: $F_{xy}(k)$ (fluxo discreto por plaquete)")
plt.colorbar(label=r"$F_{xy}$")
plt.tight_layout()
plt.show()

# (opcional) bandas ao longo de um caminho de alta simetria: Γ→X→M→Γ
def path_GXM(N=200):
    G = np.array([0.0, 0.0])
    X = np.array([np.pi, 0.0])
    M = np.array([np.pi, np.pi])
    pts = [G, X, M, G]
    kpath = []
    for A, B in zip(pts[:-1], pts[1:]):
        for t in np.linspace(0, 1, N, endpoint=False):
            kpath.append(A*(1-t) + B*t)
    kpath.append(G)
    return np.array(kpath)

kpath = path_GXM(120)
Epath = np.zeros((len(kpath), 2))
for i, (kx, ky) in enumerate(kpath):
    Epath[i] = LA.eigvalsh(Hk(kx, ky, m=m)).real

plt.figure(figsize=(7, 3.5))
plt.plot(Epath[:, 0], linewidth=1.5)
plt.plot(Epath[:, 1], linewidth=1.5)
plt.ylabel("E")
plt.title("Bandas ao longo de Γ→X→M→Γ")
plt.xticks([0, 120, 240, 360], [r"$\Gamma$", r"$X$", r"$M$", r"$\Gamma$"])
plt.tight_layout()
plt.show()
```

## Spin–orbit coupling (SOC): por que aparece e por que é importante em isolantes topológicos


Até aqui, nossos exemplos topológicos (SSH, Chern number, etc.) foram construídos a partir de fases geométricas
(Berry phase/curvature) e da estrutura de bandas. Para entender **isolantes topológicos com reversão temporal**
(Quantum Spin Hall, isolantes topológicos), precisamos adicionar um ingrediente físico central: o
**acoplamento spin–órbita (SOC)**.

Em uma frase: **SOC liga o movimento orbital do elétron ao seu spin**, e isso permite que bandas com caráter orbital
diferente se misturem de forma que preserve a reversão temporal — exatamente o cenário onde aparecem
fases topológicas do tipo $\mathbb{Z}_2$.

---

### Origem física (intuição)
O SOC é um efeito relativístico: um elétron que se move em um potencial com forte variação espacial
(por exemplo, perto de núcleos ou em ambientes com fortes campos elétricos internos) percebe, no seu
referencial, um campo magnético efetivo que depende do movimento. Esse campo efetivo acopla ao spin.

Sem entrar em uma derivação completa, o ponto essencial é:
- o SOC cresce quando o potencial varia rapidamente no espaço (tipicamente mais forte em átomos pesados);
- em sólidos, o SOC atua como um termo efetivo permitido (ou proibido) pelas **simetrias** do sistema.

---

### Dois “tipos” de SOC que aparecem em modelos efetivos
Dependendo da simetria do cristal e do confinamento, o SOC pode aparecer de maneiras diferentes.
Nesta aula vamos distinguir, conceitualmente, dois casos:

1. **SOC do tipo atômico (intrínseco):** vem da interação local $\mathbf{L}\cdot\mathbf{S}$,
   isto é, do acoplamento entre o momento angular orbital e o spin no mesmo sítio/átomo.
   Em modelos tight-binding, ele aparece como um termo on-site que mistura orbitais com fases (matrizes complexas).

2. **SOC do tipo Rashba (extrínseco):** pode surgir quando há **quebra de inversão** por um campo elétrico médio
   (interfaces, superfícies, heteroestruturas), e leva a termos que dependem explicitamente de $\mathbf{k}$ e do spin.
   (Vamos mencionar esse caso, mas o foco aqui será o SOC intrínseco, alinhado com o espírito do modelo que estudaremos.)

---

### O que o SOC muda na estrutura de bandas?
Mesmo sem interações, SOC pode:
- remover degenerescências (ou reorganizá-las de acordo com TRS);
- gerar **texturas de spin** nos autestados (o spin esperado depende de $k$);
- abrir gaps ou inverter a ordem de bandas com caráter orbital diferente.

É essa reorganização (especialmente a **inversão de bandas** induzida por SOC) que está por trás do mecanismo
mais comum para a fase Quantum Spin Hall em modelos mínimos como o BHZ (que discutiresmo aqui).

---

### Conexão com topologia ($\mathbb{Z}_2$)
Em sistemas com reversão temporal (TRS), o Chern number total associado à carga é zero,
mas ainda pode existir uma topologia robusta classificada por um invariante $\mathbb{Z}_2$.
O SOC é o ingrediente conhecido que permite realizar essa fase em materiais realistas.

Nesta aula:
- primeiro vamos construir uma intuição de SOC em um modelo 1D simples (tight-binding com dois orbitais e spin),
- em seguida conectaremos a ideia com o modelo BHZ, que será a base para calcular o invariante $\mathbb{Z}_2$.

## Simetrias com spin: reversão temporal (TRS) e inversão (P) e suas consequências em $H(\mathbf{k})$ e $E(\mathbf{k})$

Antes de entrar nos detalhes do SOC, vale fixar como as simetrias atuam quando o sistema tem **spin**.
A mensagem principal é que TRS continua relacionando $\mathbf{k}\leftrightarrow -\mathbf{k}$, mas agora a operação
de reversão temporal **não é apenas conjugação complexa**: ela também atua no espaço de spin.

---

### Reversão temporal (TRS) com spin-1/2

Para partículas com spin-1/2, a reversão temporal é uma operação antiunitária:
$$
\Theta = i\,s_y\,K,
$$
onde $s_y$ é a matriz de Pauli no espaço de spin e $K$ é conjugação complexa.

A condição de simetria de TRS no espaço recíproco é:
$$
\boxed{\Theta\,H(\mathbf{k})\,\Theta^{-1}=H(-\mathbf{k}).}
$$

**Consequência para o espectro:**
TRS implica que os autovalores aparecem em pares relacionados por $\mathbf{k}\to -\mathbf{k}$:
$$
\boxed{E_n(\mathbf{k}) = E_n(-\mathbf{k}) \ \text{(como conjunto de bandas)}.}
$$

Além disso, existe a **degenerescência de Kramers**:
nos pontos invariantes por reversão temporal (TRIM), onde $\mathbf{k}\equiv -\mathbf{k}$ módulo um vetor recíproco,
as bandas são pelo menos duplamente degeneradas:
$$
\boxed{\mathbf{k}=\Lambda_i \ \Rightarrow\ E(\Lambda_i)\ \text{é degenerado (par de Kramers)}.}
$$

Intuitivamente: TRS garante que cada estado em $\mathbf{k}$ tem um parceiro ortogonal em $-\mathbf{k}$
com o mesmo autovalor.

---

### Inversão (P)

A inversão é unitária e atua como $\mathbf{r}\to -\mathbf{r}$, portanto $\mathbf{k}\to -\mathbf{k}$.
No espaço interno (orbitais/sublattices), ela é representada por uma matriz unitária $P$ (dependente da base):
$$
\boxed{P\,H(\mathbf{k})\,P^{-1}=H(-\mathbf{k}).}
$$

**Consequência:**
assim como TRS, a inversão também força uma simetria $\mathbf{k}\leftrightarrow -\mathbf{k}$ no espectro.
A diferença é que, ao contrário de TRS, inversão **não** é antiunitária e não impõe, por si só,
degenerescência de Kramers.

---

### Quando TRS e inversão estão presentes ao mesmo tempo

Se o sistema tem **TRS e inversão**, então a combinação $P\Theta$ relaciona estados no *mesmo* $\mathbf{k}$,
e isso frequentemente implica uma degenerescência adicional (“dupla degenerescência” em muitas bandas),
além da degenerescência de Kramers nos TRIM.

Do ponto de vista prático para esta aula:
- TRS $\Rightarrow$ pares de Kramers nos TRIM.
- Inversão $\Rightarrow$ permite classificar estados nos TRIM por **paridade** ($\pm 1$).
- A combinação das duas é o que torna possível um cálculo muito simples do invariante $\mathbb{Z}_2$
  em modelos com inversão (método de Fu–Kane, via paridades nos TRIM).

---

### Uma regra prática (boa para diagnosticar no notebook)

- Quebrar TRS (por exemplo com um termo de Zeeman) pode remover a degenerescência de Kramers nos TRIM.
- Quebrar inversão (por exemplo com termos A/B diferentes, ou assimetrias estruturais) pode permitir
  efeitos como “splittings” dependentes de $\mathbf{k}$ e texturas de spin mais ricas, mesmo mantendo TRS.
- Se TRS é preservada, o espectro continua satisfazendo $E(\mathbf{k})=E(-\mathbf{k})$ como conjunto de bandas,
  mas os autovetores em $\mathbf{k}$ e $-\mathbf{k}$ são relacionados por $\Theta$ (e têm spins opostos na linguagem usual).

Essas relações serão úteis para entender por que o Quantum Spin Hall é compatível com TRS e por que seu
invariante topológico é $\mathbb{Z}_2$ (e não um Chern number de carga).

# Do QWZ (Chern insulator) ao QSH/BHZ (duas cópias TRS)

## A) O que acabamos de fazer: QWZ é um Chern insulator (sem TRS)
No modelo de 2 bandas (QWZ),
$$
H(\mathbf{k}) = d_x(\mathbf{k})\,\sigma_x + d_y(\mathbf{k})\,\sigma_y + d_z(\mathbf{k})\,\sigma_z,
$$
a banda ocupada pode ter um número de Chern
$$
C=\frac{1}{2\pi}\int_{\mathrm{BZ}} d^2k\,\Omega(\mathbf{k}),
$$
e quando $C\neq 0$ o sistema tem **bordas quirais** (bulk–boundary correspondence).

**O que “borda quiral” quer dizer na prática:**
- Considere uma amostra com contorno (por exemplo, uma fita: periódica em $x$ e aberta em $y$).
- O bulk é isolante (tem gap), mas na borda aparece um ramo de estados que **atravessa o gap**.
- *Quiral* significa que esse ramo tem **um único sentido de propagação** em uma dada borda: a velocidade de grupo
  $v=\partial E/\partial k_x$ tem sinal fixo para o modo que vive naquela borda.
- Em linguagem de transporte: a borda funciona como uma **“via de mão única”**. Como não há um canal contra-propagante
  na mesma borda, desordem não-magnética não consegue causar retroespalhamento (backscattering) elástico de forma eficiente.
  Isso explica a robustez do modo.

**Mensagem:** um único bloco $2\times 2$ com $C\neq 0$ já descreve um isolante topológico do tipo “Chern insulator”,
com corrente de borda unidirecional.

---

## B) Agora: Quantum Spin Hall (QSH) = duas cópias relacionadas por TRS
Para falar de isolantes topológicos com reversão temporal (TRS), precisamos incluir spin.
A construção mínima é:
- um bloco para “spin up” com Hamiltoniano $h(\mathbf{k})$,
- e um bloco para “spin down” que é a cópia imposta por TRS: $h^*(-\mathbf{k})$.

Assim, escrevemos um Hamiltoniano $4\times 4$ bloco-diagonal:
$$
H(\mathbf{k})=
\begin{pmatrix}
h(\mathbf{k}) & 0 \\
0 & h^*(-\mathbf{k})
\end{pmatrix}.
$$

Se o bloco de cima tem Chern $C_\uparrow=C$, o bloco TRS tem
$$
C_\downarrow=-C_\uparrow.
$$
Logo, o Chern total de carga é
$$
C_{\mathrm{tot}} = C_\uparrow + C_\downarrow = 0,
$$
como deve ser em um sistema com TRS (não há Hall de carga).

**Mas a borda muda de natureza: bordas helicais.**
- Agora não existe “mão única”: TRS obriga a existência de pares.
- Em cada borda aparecem **dois modos contra-propagantes**, que formam um **par de Kramers**.
- No modelo ideal em que $S_z$ é conservado, podemos pensar que um modo é “$\uparrow$ para a direita” e o outro é
  “$\downarrow$ para a esquerda” (na prática, pode ser uma textura e não um spin fixo).
- A proteção é diferente do caso quiral: aqui existe um canal contra-propagante, mas TRS impede que um estado espalhe
  elasticamente para o seu parceiro de Kramers via desordem **não magnética**. Em termos simples: para retroespalhar,
  seria preciso acoplar dois estados relacionados por TRS, o que é proibido enquanto TRS for preservada.

**Frase para guardar:**
> QSH é como “dois Chern insulators com Chern oposto”, costurados por TRS: o Hall de carga cancela, mas ficam bordas helicais.

---

## C) Onde entra SOC (sem entrar em detalhes técnicos)
Em materiais reais, o spin não aparece “do nada”: o acoplamento spin–órbita (SOC) é o mecanismo que reorganiza e mistura
bandas com diferentes caracteres orbitais, permitindo inversão de bandas e um gap topológico.

No modelo BHZ (que descreve QSH de forma mínima realista), a estrutura “duas cópias TRS” aparece naturalmente (pares de Kramers),
e a fase topológica é controlada por uma inversão de bandas (mudança efetiva do sinal de uma “massa” em pontos de alta simetria).

---

## D) Spin Chern e o invariante $\mathbb{Z}_2$ (comentário operacional)
Se o Hamiltoniano é bloco-diagonal em spin (isto é, $S_z$ é conservado), então faz sentido falar no Chern de cada bloco:
$$
C_\uparrow,\quad C_\downarrow=-C_\uparrow.
$$
Nesse caso, o invariante $\mathbb{Z}_2$ pode ser obtido simplesmente por
$$
\nu = (C_\uparrow-C_\downarrow) \bmod 2 \in \{0,1\}.
$$

- fase QSH ($\nu=1$): bordas helicais protegidas por TRS.
- fase trivial ($\nu=0$).

(Se $S_z$ não é conservado, ainda existe $\nu$, mas o cálculo é mais sofisticado; aqui só vamos comentar isso.)

####Extra: Fukui no modelo de Haldane (rede honeycomb)

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Lattice geometry (your NN vectors)
# -----------------------------
def lattice_vectors(a=1.0):
    # Your NN vectors:
    d1 = np.array([0.0, a])
    d2 = np.array([np.sqrt(3)/2*a, -0.5*a])
    d3 = np.array([-np.sqrt(3)/2*a, -0.5*a])
    deltas = np.stack([d1, d2, d3], axis=0)

    # Primitive Bravais vectors for triangular lattice (A sublattice):
    # a1 = d2 - d3, a2 = d1 - d3, a3 = a2 - a1
    a1 = d2 - d3                      # (sqrt(3)a, 0)
    a2 = d1 - d3                      # (sqrt(3)/2 a, 3/2 a)
    a3 = a2 - a1                      # (-sqrt(3)/2 a, 3/2 a)
    avecs = np.stack([a1, a2, a3], axis=0)

    return deltas, avecs, a1, a2

def reciprocal_vectors(a1, a2):
    # Solve for b1, b2 such that a_i · b_j = 2π δ_ij
    A = np.stack([a1, a2], axis=1)          # columns are a1,a2 (2x2)
    B = 2*np.pi * np.linalg.inv(A).T        # columns are b1,b2 (2x2)
    b1 = B[:, 0]
    b2 = B[:, 1]
    return b1, b2

def k_grid_uv(N1, N2, b1, b2):
    # Rectangular grid in (u,v) torus: k = u b1 + v b2
    k = np.zeros((N1, N2, 2), dtype=float)
    for m in range(N1):
        u = m / N1
        for n in range(N2):
            v = n / N2
            k[m, n] = u*b1 + v*b2
    return k

# -----------------------------
# Haldane model Bloch Hamiltonian (2x2)
# H(k) = d0(k) I + d(k) · σ
# Off-diagonal from NN hopping t1 * f(k), f = Σ_j e^{i k·δ_j}
# Diagonal terms from:
#   sublattice mass M (σ_z)
#   complex NNN hopping t2 e^{± i phi} -> d0 and d_z contributions using a1,a2,a3
# -----------------------------
def haldane_Hk(kx, ky, deltas, avecs, t1=1.0, t2=0.1, phi=np.pi/2, M=0.0):
    k = np.array([kx, ky])

    # NN structure factor
    f = np.sum(np.exp(1j * (deltas @ k)))

    # NNN sums using Bravais vectors a1,a2,a3 (same-sublattice hops)
    # Standard Haldane form:
    # d0 = 2 t2 cos(phi) * Σ_i cos(k·a_i)
    # dZ = M - 2 t2 sin(phi) * Σ_i sin(k·a_i)
    ka = avecs @ k
    csum = np.sum(np.cos(ka))
    ssum = np.sum(np.sin(ka))

    d0 = 2.0 * t2 * np.cos(phi) * csum
    dZ = M - 2.0 * t2 * np.sin(phi) * ssum

    # Convention: H_AB = t1 f, H_BA = t1 f*
    # Then dx,dy correspond to Re/Im of f, but we can just build H directly.
    H = np.array([[d0 + dZ,       t1 * f],
                  [t1 * np.conj(f), d0 - dZ]], dtype=complex)
    return H

# -----------------------------
# Fukui method (Abelian) for lowest band
# Returns: Chern number, plaquette flux F12 (shape (N1,N2))
# -----------------------------
def fukui_chern_and_flux(N1=41, N2=41, a=1.0, t1=1.0, t2=0.1, phi=np.pi/2, M=0.0):
    deltas, avecs, a1, a2 = lattice_vectors(a=a)
    b1, b2 = reciprocal_vectors(a1, a2)

    k = k_grid_uv(N1, N2, b1, b2)

    # eigenvectors for lowest band
    u = np.zeros((N1, N2, 2), dtype=complex)

    for m in range(N1):
        for n in range(N2):
            kx, ky = k[m, n]
            H = haldane_Hk(kx, ky, deltas, avecs, t1=t1, t2=t2, phi=phi, M=M)
            evals, evecs = np.linalg.eigh(H)  # columns are eigenvectors
            idx0 = np.argmin(evals)           # lowest band
            vec = evecs[:, idx0]
            # gauge choice doesn't matter for Fukui; but normalize to be safe:
            vec = vec / (np.linalg.norm(vec) + 1e-30)
            u[m, n] = vec

    # link variables U1, U2 (unit modulus complex numbers)
    U1 = np.zeros((N1, N2), dtype=complex)
    U2 = np.zeros((N1, N2), dtype=complex)

    def link(v1, v2):
        ov = np.vdot(v1, v2)               # <v1|v2>
        return ov / (np.abs(ov) + 1e-30)   # phase only

    for m in range(N1):
        mp = (m + 1) % N1
        for n in range(N2):
            np_ = (n + 1) % N2
            U1[m, n] = link(u[m, n], u[mp, n])
            U2[m, n] = link(u[m, n], u[m, np_])

    # plaquette Berry flux (discrete curvature)
    F12 = np.zeros((N1, N2), dtype=float)

    for m in range(N1):
        mp = (m + 1) % N1
        for n in range(N2):
            np_ = (n + 1) % N2

            # Fukui plaquette variable
            W = U1[m, n] * U2[mp, n] / (U1[m, np_] * U2[m, n])
            F12[m, n] = np.angle(W)  # principal branch (-pi, pi]

    C = np.sum(F12) / (2*np.pi)

    # k-space area of one plaquette (for plotting curvature density)
    area_BZ = abs(np.cross(b1, b2))        # scalar in 2D (|b1 x b2|)
    dA = area_BZ / (N1 * N2)

    return C, F12, dA, (b1, b2), (a1, a2)

# -----------------------------
# Run example + plots
# -----------------------------
if __name__ == "__main__":
    # Parameters (typical topological point: M=0, phi=pi/2 gives C=±1 depending on sign of t2*sin(phi))
    a = 1.0
    t1 = 1.0
    t2 = 0.0
    phi = np.pi/2
    M = 0.5

    N1, N2 = 61, 61

    C, F12, dA, (b1, b2), (a1, a2) = fukui_chern_and_flux(
        N1=N1, N2=N2, a=a, t1=t1, t2=t2, phi=phi, M=M
    )

    print("a1 =", a1, " a2 =", a2)
    print("b1 =", b1, " b2 =", b2)
    print("Chern number (Fukui) =", C, " -> rounded:", int(np.rint(C)))

    # Berry curvature density approx: Omega(k) ~ flux / plaquette area
    Omega = F12 / dA

    # Plot: flux per plaquette (gauge-invariant)
    plt.figure()
    plt.imshow(F12.T, origin="lower", aspect="auto")
    plt.colorbar(label="plaquette flux  F12  (radians)")
    plt.xlabel("m (u-direction)")
    plt.ylabel("n (v-direction)")
    plt.title("Fukui Berry flux per plaquette (Haldane)")

    # Plot: curvature density
    plt.figure()
    plt.imshow(Omega.T, origin="lower", aspect="auto")
    plt.colorbar(label="Berry curvature density  Ω(k) ≈ F12/dA")
    plt.xlabel("m (u-direction)")
    plt.ylabel("n (v-direction)")
    plt.title("Berry curvature density on (u,v) rectangular BZ grid")

    plt.show()
```
